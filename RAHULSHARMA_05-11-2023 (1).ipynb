```python
##1.1 What is the probability that a randomly chosen player would suffer an injury?

total_injured = 145
total_non_injured = 90
total_players = total_injured + total_non_injured

# Calculate the probability of injury
probability_injury = total_injured / total_players

# Display the probability as a percentage
probability_percentage = probability_injury * 100

print(f"The probability that a randomly chosen player would suffer an injury is {probability_percentage:.2f}%")
```

```python
##1.2 What is the probability that a player is a forward or a winger?

# Given data
total_forward = 56
total_winger = 20
total_players = total_forward + total_winger

# Calculate the probability that a player is a forward or a winger
probability_forward_or_winger = total_players / (total_forward + total_winger)

# Display the probability as a percentage
probability_percentage = probability_forward_or_winger * 100

print(f"The probability that a player is a forward or a winger is {probability_percentage:.2f}%")
```

```python
##1.3 What is the probability that a randomly chosen player plays in a striker position and has a foot injury?

# Given data
injured_striker = 45
total_players = 145  # Total injured players

# Calculate the probability of being a striker with a foot injury
probability_striker_with_injury = injured_striker / total_players

# Display the probability as a percentage
probability_percentage = probability_striker_with_injury * 100

print(f"The probability that a randomly chosen player plays in a striker position and has a foot injury is {probability_percentage:.2f}%")
```

```python
##1.4 What is the probability that a randomly chosen injured player is a striker?

# Given data
injured_striker = 45
total_injured = 145  # Total injured players

# Calculate the probability that a randomly chosen injured player is a striker
probability_injured_striker = injured_striker / total_injured

# Display the probability as a percentage
probability_percentage = probability_injured_striker * 100

print(f"The probability that a randomly chosen injured player is a striker is {probability_percentage:.2f}%")
```

```python
##2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

import scipy.stats as stats

mean = 5
std_dev = 1.5
value = 3.17

# Calculate the z-score
z = (value - mean) / std_dev

# Create a range of values for the x-axis
x = np.linspace(0, 10, 1000)

# Calculate the cumulative distribution function (CDF) for the range of x values
cdf = stats.norm.cdf(x, loc=mean, scale=std_dev)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, cdf, color='blue')
plt.fill_between(x, 0, cdf, where=(x <= value), color='gray', alpha=0.5)
plt.title('Cumulative Distribution of Breaking Strength')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()
```

```python
##2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

mean = 5
std_dev = 1.5
value = 3.6

# Calculate the z-score
z = (value - mean) / std_dev

# Find the probability using the cumulative distribution function (CDF)
probability = 1 - stats.norm.cdf(z)

# Create the plot
x = np.linspace(0, 10, 1000)
cdf = stats.norm.cdf(x, loc=mean, scale=std_dev)

plt.figure(figsize=(8, 6))
plt.plot(x, cdf, color='blue')
plt.fill_between(x, 0, cdf, where=(x >= value), color='gray', alpha=0.5)
plt.title('Cumulative Distribution of Breaking Strength')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()

print(f"The proportion of gunny bags with breaking strength of at least 3.6 kg per sq. cm is {probability:.4f}")
```

```python
##2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

mean = 5
std_dev = 1.5
value1 = 5
value2 = 5.5

# Calculate the z-scores
z1 = (value1 - mean) / std_dev
z2 = (value2 - mean) / std_dev

# Find the probabilities
probability1 = stats.norm.cdf(z1)
probability2 = stats.norm.cdf(z2)

# Calculate the proportion
proportion = probability2 - probability1

# Create the plot
x = np.linspace(0, 10, 1000)
cdf = stats.norm.cdf(x, loc=mean, scale=std_dev)

plt.figure(figsize=(8, 6))
plt.plot(x, cdf, color='blue')
plt.fill_between(x, 0, cdf, where=((x >= value1) & (x <= value2)), color='gray', alpha=0.5)
plt.title('Cumulative Distribution of Breaking Strength')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()

print(f"The proportion of gunny bags with breaking strength between 5 and 5.5 kg per sq. cm is {proportion:.4f}")
```

```python
##2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

mean = 5
std_dev = 1.5
lower_limit = 3
upper_limit = 7.5

# Calculate the z-scores for the lower and upper limits
z_lower = (lower_limit - mean) / std_dev
z_upper = (upper_limit - mean) / std_dev

# Find the probabilities
probability_lower = stats.norm.cdf(z_lower)
probability_upper = stats.norm.cdf(z_upper)

# Calculate the proportion outside the range
proportion_outside_range = 1 - (probability_upper - probability_lower)

# Create the plot
x = np.linspace(0, 10, 1000)
cdf = stats.norm.cdf(x, loc=mean, scale=std_dev)

plt.figure(figsize=(8, 6))
plt.plot(x, cdf, color='blue')
plt.fill_between(x, 0, cdf, where=((x < lower_limit) | (x > upper_limit)), color='gray', alpha=0.5)
plt.title('Cumulative Distribution of Breaking Strength')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Cumulative Probability')
plt.grid()
plt.show()

print(f"The proportion of gunny bags with breaking strength NOT between 3 and 7.5 kg per sq. cm is {proportion_outside_range:.4f}")
```

```python
##3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

import pandas as pd

# Read the CSV file
data = pd.read_csv('Zingaro_Company.csv')

# Extract the 'Unpolished' column
unpolished_data = data['Unpolished ']

# Calculate statistics
mean_unpolished = unpolished_data.mean()
std_unpolished = unpolished_data.std()

# Define printing requirements (adjust these values as needed)
required_mean = 150  # Example mean requirement
required_std = 10    # Example standard deviation requirement

# Check if unpolished stones meet the printing requirements
if mean_unpolished >= required_mean and std_unpolished <= required_std:
    print("The unpolished stones meet the printing requirements.")
else:
    print("The unpolished stones may not meet the printing requirements.")

# You can also perform visual inspections, tests, and consult experts as needed.
```

```python
##3.2 Is the mean hardness of the polished and unpolished stones the same?

import pandas as pd
from scipy.stats import ttest_ind



# Extract the 'Treated and Polished' and 'Unpolished' columns
polished_data = data['Treated and Polished']
unpolished_data = data['Unpolished ']

# Perform a t-test
t_statistic, p_value = ttest_ind(polished_data, unpolished_data)

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha (significance level)
if p_value < alpha:
    print("The mean hardness of polished and unpolished stones is significantly different.")
else:
    print("There is no significant difference in the mean hardness of polished and unpolished stones.")
```

```python
##4.1 How does the hardness of implants vary depending on dentists?

import pandas as pd

# Load the Excel file into a DataFrame
df = pd.read_excel('Dental+Hardness+data.xlsx', sheet_name='Sheet1')


# Create a box plot using the 'Response' column
sns.boxplot(x='Dentist', y='Response', data=df)
plt.xlabel('Dentist')
plt.ylabel('Response (Hardness)')
plt.title('Hardness by Dentist')
plt.show()
```

```python
##4.2 How does the hardness of implants vary depending on methods?

from scipy.stats import f_oneway

# Perform one-way ANOVA to compare hardness measurements by method
method_groups = [group['Response'] for name, group in df.groupby('Method')]
f_statistic, p_value = f_oneway(*method_groups)

print("ANOVA F-statistic:", f_statistic)
print("ANOVA p-value:", p_value)


import matplotlib.pyplot as plt
import seaborn as sns

# Create visualizations (e.g., box plots, histograms)
sns.boxplot(x='Method', y='Response', data=df)
plt.xlabel('Method')
plt.ylabel('Response (Hardness)')
plt.title('Hardness by Method')
plt.show()


```

```python
###4.3 What is the interaction effect between the dentist and method on the hardness of dental implants for each type of alloy?

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Perform a two-way ANOVA
formula = 'Response ~ Dentist * Method * Alloy'
model = ols(formula, df).fit()
anova_results = anova_lm(model)

print(anova_results)
```

```python
###4.4 How does the hardness of implants vary depending on dentists and methods together?

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Perform a two-way ANOVA
formula = 'Response ~ Dentist * Method'
model = ols(formula, df).fit()
anova_results = anova_lm(model)

print(anova_results)
```
